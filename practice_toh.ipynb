```python
def hanoi(n, a, b, c):
    count = 1
    if n == 1:
        print("move {} from {} to {}".format(n, a, c))
        return count
    else:
        count += hanoi(n - 1, a, c, b)
        print("move {} from {} to {}".format(n, a, c))
        count += hanoi(n - 1, b, a, c)
        return count
```

```python
a = 'A'
b = 'B'
c = 'C'

n = int(input("Enter the number of rings: "))
count = hanoi(n, a, b, c)
print("Total no of moves: ", count)
```
