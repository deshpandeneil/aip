```python
def bfs(start, end, x_max, y_max):
    front = []
    path = []
    visited = []
    
    front.append(start)
    while(len(front) > 0):
        curr = front.pop()
        x = curr[0]
        y = curr[1]
        path.append(curr)
        
        if x == end or y == end:
            return path
        
        #rules
        
        if curr[0] < x_max and [x_max, curr[1]] not in visited:
            front.append([x_max, curr[1]])
            visited.append([x_max, curr[1]])
        
        if curr[1] < y_max and [curr[0], y_max] not in visited:
            front.append([curr[0], y_max])
            visited.append([curr[0], y_max])
            
        if curr[0] > x_max and [0, curr[1]] not in visited:
            front.append([0, curr[1]])
            visited.append([0, curr[1]])
            
        if curr[1] > y_max and [x_max, 0] not in visited:
            front.append([x_max, 0])
            visited.append([x_max, 0])
            
        if curr[0] > 0 and [max(0, x + y - y_max), min(x + y, y_max)] not in visited:
            front.append([max(0, x + y - y_max), min(x + y, y_max)])
            visited.append([max(0, x + y - y_max), min(x + y, y_max)])
        
        if curr[1] > 0 and [min(x + y, x_max), max(0, x + y - x_max)] not in visited:
            front.append([min(x + y, x_max), max(0, x + y - x_max)])
            visited.append([min(x + y, x_max), max(0, x + y - x_max)])
    return "Not found"
```

```python
x_max = int(input("Enter capacity of jug 1: "))
y_max = int(input("Enter capacity of jug 2: "))
end = int(input("Enter target quantity: "))

start = [0, 0]

bfs(start, end, x_max, y_max)
```
