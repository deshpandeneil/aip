```python
import copy
```

```python
inf = float('inf')
```

```python
class TSP:
    def __init__(self, city_matrix, source):
        self.city_matrix = city_matrix
        self.n = len(city_matrix)
        self.source = source
    
    def solve(self):
        minCost = inf
        for i in range(self.n):
            cost = self._solve(copy.deepcopy(city_matrix), i, i)
            if cost and cost < minCost:
                minCost = cost
        return minCost
    
    def _solve(self, city_matrix, current, source):
        for i in range(self.n):
            city_matrix[i][current] = inf
        curMin = inf
        index = 0
        for i in range(self.n):
            if(city_matrix[current][i] < curMin):
                curMin = city_matrix[current][i]
                index = i
        if curMin == inf:
            return self.city_matrix[current][source]
        city_matrix[current][index] = inf
        city_matrix[index][current] = inf
        return curMin + self._solve(city_matrix, index, source)
```

```python
if __name__ == '__main__':
    city_matrix = [
        [inf, 10, 15, 20],
        [10, inf, 35, 25],
        [15, 35, inf, 30],
        [20, 25, 30, inf],
    ]
    source = 0
    tsp = TSP(city_matrix, source)
    cost = tsp.solve()
    print(cost)
```
